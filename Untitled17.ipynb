```python
import zipfile

with zipfile.ZipFile('archive (5).zip', 'r') as zip_ref:
    zip_ref.extractall('ipl_data')

print("ZIP Extracted Successfully")
```

```python
import pandas as pd

df = pd.read_csv('ipl_data/IPL.csv')

print(df.shape)
print(df.columns)

df.head()
```

```python
# Create batting statistics

batting = df.groupby('batter').agg(
    Runs=('batter_runs', 'sum'),
    Balls=('ball', 'count'),
    Matches=('match_id', 'nunique')
).reset_index()

# Remove players with very few matches
batting = batting[batting['Matches'] >= 10]

# Strike Rate
batting['Strike_Rate'] = (
    batting['Runs'] / batting['Balls']
) * 100

# Batting Impact Score
batting['Batting_Impact'] = (
    batting['Runs'] * 0.6 +
    batting['Strike_Rate'] * 0.4
)

# Top 20 batsmen
top_batsmen = batting.sort_values(
    'Batting_Impact',
    ascending=False
)

top_batsmen.head(20)
```

```python
# Bowling Statistics

bowling = df.groupby('bowler').agg(
    Wickets=('bowler_wicket', 'sum'),
    Balls=('ball', 'count'),
    Matches=('match_id', 'nunique')
).reset_index()

# Bowlers with at least 10 matches
bowling = bowling[bowling['Matches'] >= 10]

# Bowling Impact Score
bowling['Bowling_Impact'] = (
    bowling['Wickets'] * 10
)

# Top Bowlers
top_bowlers = bowling.sort_values(
    'Bowling_Impact',
    ascending=False
)

top_bowlers.head(20)
```

```python
# Merge batting and bowling data

talent = pd.merge(
    batting[['batter', 'Batting_Impact']],
    bowling[['bowler', 'Bowling_Impact']],
    left_on='batter',
    right_on='bowler',
    how='outer'
)

# Fill missing values
talent['Batting_Impact'] = talent['Batting_Impact'].fillna(0)
talent['Bowling_Impact'] = talent['Bowling_Impact'].fillna(0)

# Create player column
talent['Player'] = talent['batter'].fillna(talent['bowler'])

# Final Talent Score
talent['Talent_Score'] = (
    talent['Batting_Impact'] * 0.6 +
    talent['Bowling_Impact'] * 0.4
)

# Rank players
talent = talent[['Player', 'Batting_Impact', 'Bowling_Impact', 'Talent_Score']]

top_players = talent.sort_values(
    'Talent_Score',
    ascending=False
)

top_players.head(20)
```

```python
import matplotlib.pyplot as plt

top10 = top_players.head(10)

plt.figure(figsize=(12,6))
plt.bar(top10['Player'], top10['Talent_Score'])
plt.xticks(rotation=45)
plt.title('Top 10 Cricket Talents')
plt.xlabel('Player')
plt.ylabel('Talent Score')
plt.show()
```

```python
top_players.to_csv('Cricket_Talent_Rankings.csv', index=False)

print("File Saved Successfully")
```
